```python
import pandas as pd
data = pd.read_csv("finalmovie.csv")
data = data.sort_values(by = ['release_date']) 
data = data[800:]
data.head()
```

```python
nozero = data[data["budget"] != 0]
nozero = nozero[nozero["revenue"] != 0]
nozero.head(20)
```

```python
df = nozero
trace_high = go.Scatter(
                x=df.release_date,
                y=df['revenue'],
                name = "Revenue",
                line = dict(color = '#17BECF'),
                opacity = 0.8)

trace_low = go.Scatter(
                x=df.release_date,
                y=df['budget'],
                name = "Budget",
                line = dict(color = '#7F7F7F'),
                opacity = 0.8)

data = [trace_high,trace_low]


layout = dict(
    title='Revenue and Budget Changing in Years',
    xaxis=dict(title = "years",
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label='1m',
                     step='month',
                     stepmode='backward'),
                dict(count=6,
                     label='6m',
                     step='month',
                     stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(
            visible = True
        ),
        type='date'
    ),
    yaxis = dict(title = "Revenue")
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename = "Time Series with Rangeslider")
```

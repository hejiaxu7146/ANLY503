```python
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt
import ast
```

```python
df = pd.read_csv("finalmovie.csv")
df.head()
```

```python
dfaction = df[df["genres"] == "Action"]
```

```python
actionwords = []
texts = " "
for w in dfaction["keywords"]:
    w = ast.literal_eval(w)
    words = (' ').join(w)
    texts = texts + words
actionwordcloud = WordCloud().generate(texts)
plt.imshow(actionwordcloud, interpolation='bilinear')
plt.title("WordCloud for Action Movies")
plt.axis("off")
```
